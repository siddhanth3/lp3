## Gradient Descent Algorithm

#### find the local minima of the function y=(x+3)Â² starting from the point x=2

```python
# Import required library
import matplotlib.pyplot as plt
```

```python

# Function and its derivative
def f(x):
    return (x + 3)**2   # Function: y = (x + 3)^2

def f_prime(x):
    return 2*(x + 3)    # Derivative: dy/dx = 2(x + 3)
```

```python
# Initialize parameters
x = 2           # Starting point
learning_rate = 0.1
epochs = 25     # Number of iterations
```

```python
# To store results
x_values = [x]
y_values = [f(x)]
```

```python

# Gradient Descent loop
for i in range(epochs):
    grad = f_prime(x)          # Compute gradient
    x = x - learning_rate * grad  # Update x value
    x_values.append(x)
    y_values.append(f(x))
    print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
```

```python
# Final result
print("\nLocal minimum occurs at x =", round(x, 4))
print("Minimum value of function =", round(f(x), 4))
```

```python
# Plotting the function and the path taken
plt.figure(figsize=(8,5))
x_plot = [i for i in range(-6, 4)]
y_plot = [f(i) for i in x_plot]
plt.plot(x_plot, y_plot, label="y = (x + 3)^2", color='blue')
plt.scatter(x_values, y_values, color='red', label="Descent steps")
plt.title("Gradient Descent on y = (x + 3)^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
